```python
#Drop unwanted factors
creditData.drop(['telephone','residence_history','dependents'],axis=1, inplace=True)
```

```python
#Enrich the data
creditData['checking_balance'].replace('unknown', 'no account',inplace = True)
creditData['checking_balance'].replace('< 0 DM', 'no balance', inplace = True)
creditData['savings_balance'].replace('unknown', 'no account', inplace = True)
creditData['savings_balance'].replace('< 0 DM', 'no balance', inplace = True)
creditData['savings_balance'].replace(['501 - 1000 DM','> 1000 DM'], '>500 DM', inplace = True)
```

```python
creditData['credit_history'].replace(['fully repaid this bank', 'fully repaid'], 'repaid', inplace = True)
```

```python
creditData['purpose'].replace(['furniture', 'repairs','domestic appliances','radio/tv'], 'home related', inplace = True)
creditData['purpose'].replace('car (new)', 'new car', inplace = True)
creditData['purpose'].replace('car (used)', 'used car', inplace = True)
```

```python
creditData['employment_length'].replace('0 - 1 yrs', 'Entry-level', inplace = True)
creditData['employment_length'].replace('1 - 4 yrs', 'mid-level', inplace = True)
creditData['employment_length'].replace('4 - 7 yrs', 'mid-level', inplace = True)
creditData['employment_length'].replace('> 7 yrs', 'senior-level', inplace = True)
```

```python
creditData['other_debtors'].replace('none', 'no', inplace = True)
creditData['other_debtors'].replace(['guarantor','co-applicant'], 'yes', inplace = True)
creditData.rename(columns = {'other_debtors':'guarantor'}, inplace = True)
```

```python
creditData['installment_plan'].replace(['bank','stores'], 'bank/stores', inplace = True)
```

```python
creditData['job'].replace(['unskilled resident','unemployed non-resident'], 'unskilled/unemployed', inplace = True)
```

```python
import numpy as np
creditData['default'] = np.where(creditData['default'] == 1, 'no', 'yes')
```

```python
from sklearn.datasets import fetch_20newsgroups_vectorized
from sklearn.feature_selection import chi2
from scipy.stats import chi2_contingency

categoricalColumns=creditData.select_dtypes(include=['object']).columns.tolist()
pvalue_list=[]
for i in categoricalColumns:
    chisqt=pd.crosstab(creditData['default'],creditData[i], margins=True)
    value = np.array([chisqt.iloc[0][0:5].values,chisqt.iloc[1][0:5].values])
    pvalue_list.append(chi2_contingency(value)[1])

pvalues= zip(categoricalColumns,pvalue_list)
pvalues=[*pvalues]
significant_variables= [(name, pvalue) for name,pvalue in pvalues if pvalue<0.05]
significant_variables
```

```python
categoricalColumns=creditData.select_dtypes(include=['object']).columns.tolist()
one_hot_encoded_data = pd.get_dummies(creditData, columns = categoricalColumns)
(one_hot_encoded_data)
```

```python
target_variable= one_hot_encoded_data['default_yes']
```

```python
independent_variable= one_hot_encoded_data.copy()
```

```python
independent_variable.drop(['default_yes', 'default_no'], axis=1, inplace=True)
```

```python
independent_variable
```

```python
import statsmodels.api as sm
logit_model=sm.Logit(target_variable,independent_variable)
result=logit_model.fit()
print(result.summary())
```

```python
one_hot_encoded_data.drop('default_no',axis=1, inplace=True)
```

```python
#Use sklearn for splitting data into test and train.
from sklearn.model_selection import train_test_split
training_data, testing_data = train_test_split(one_hot_encoded_data, test_size=0.2, random_state=25)
```

```python
training_data
```

```python
from sklearn import linear_model
X_train= training_data.drop('default_yes', axis=1)
Y_train= training_data['default_yes']
X_test= testing_data.drop('default_yes', axis=1)
Y_test= testing_data['default_yes']
logr = linear_model.LogisticRegression()
logr.fit(X_train,Y_train)
```

```python
y_pred = logr.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logr.score(X_test, Y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix_imbalanced_data = confusion_matrix(Y_test, y_pred)
print(confusion_matrix_imbalanced_data)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(Y_test, logr.predict(X_test))
fpr, tpr, thresholds = roc_curve(Y_test, logr.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

# Used L2 regularization to increase score

```python
from sklearn.model_selection import train_test_split
selectedVariablesData=creditData[['checking_balance','credit_history','purpose','savings_balance','employment_length','personal_status','property','installment_plan','housing','default','foreign_worker','months_loan_duration','installment_rate','amount']]
selectedCategricalColumns=['checking_balance','credit_history','purpose','savings_balance','employment_length','personal_status','property','installment_plan','housing','default','foreign_worker']
one_hot_encoded_selectedVariablesData = pd.get_dummies(selectedVariablesData, columns = selectedCategricalColumns)
training_data, testing_data = train_test_split(one_hot_encoded_selectedVariablesData, test_size=0.2, random_state=25)
```

```python
X_train_selectedData= training_data.drop(['default_yes','default_no'], axis=1)
Y_train_selectedData= training_data['default_yes']
X_test_selectedData= testing_data.drop(['default_yes','default_no'], axis=1)
Y_test_selectedData= testing_data['default_yes']
```

```python
logr = linear_model.LogisticRegression(penalty='l2', solver='liblinear')
logr.fit(X_train_selectedData,Y_train_selectedData)
y_pred_selectedData = logr.predict(X_test_selectedData)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logr.score(X_test_selectedData, Y_test_selectedData)))
confusion_matrix_imbalanced_data = confusion_matrix(Y_test_selectedData, y_pred_selectedData)
print(confusion_matrix_imbalanced_data)
print(classification_report(Y_test_selectedData, y_pred_selectedData))
```
